```python
import matplotlib.pyplot as plt

x = range(1,13)
y = range(10,130,10)
plt.bar(x,y)
plt.show()
```

```python
x = range(1,13)
y = range(10,130,10)
plt.bar(x, y, width = 1, align = 'edge', color = 'red', edgecolor = 'blue', linewidth = 2)
plt.show()
```

```python
x = range(1,13)
y = range(10,130,10)

plt.figure(figsize=(20,10))
plt.bar(x,y)

plt.title('Monthly expenses')
plt.xlabel('Month')
plt.ylabel('Cost($)')
plt.legend(['cost'])
plt.grid(True, axis = 'y')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3)
fig.set_size_inches(16,8)
axs[0][0].set(title = '1st')
axs[0][1].set(title = '2nd')
axs[0][2].set(title = '3rd')
axs[1][0].set(title = '4th')
axs[1][1].set(title = '5th')
axs[1][2].set(title = '6th')
axs[1][2].set(xlabel = 'month', ylabel = 'cost', title = '6th plot')
axs[1][2].bar(x,y)
```

```python
plt.rcParams['font.family'] = 'NanumGothic'  # 한글 쓰고싶을 때
plt.rcParams['axes.unicode_minus'] = False   # 마이너스 기호 쓰고싶을 때
```

```python
import numpy as np
x = np.arange(2015, 2021)
y = np.arange(len(x))
plt.bar(x,y)
plt.xticks(rotation = 45)   # x축 이름 기울였음
plt.show()
```

```python
df = sns.load_dataset('tips')

fig, ax = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(16,8)

sns.scatterplot(x = df['total_bill'], y = df['tip'], ax = ax[0][0])
sns.barplot(x = df['size'], y = df['tip'], ax = ax[0][1])
sns.regplot(x = df['total_bill'], y = df['tip'], ax = ax[1][0])
sns.violinplot(x = df['size'], y = df['tip'], ax =ax[1][1])
```
