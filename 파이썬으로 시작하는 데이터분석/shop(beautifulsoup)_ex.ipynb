```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
base_url = 'https://superstari.co.kr'
shoe_total_url = '/shop/shopbrand.html?xcode=103'
base_url + shoe_total_url
```

```python
response = requests.get(base_url + shoe_total_url)
soup = BeautifulSoup(response.content, 'lxml') # 영상과 이미지 많아서 text 보다 content
soup
```

```python
shoe_info = soup.findAll('dd', {'class' : 'prd-info'})
shoe_info[0]
```

```python
shoe_url = shoe_info[0].find('a').get('href')
print(shoe_url)

name = shoe_info[0].find('dt',{'class' : 'prd-name'}).text
print(name)

price = shoe_info[0].find('span',{'class' : 'check_price_sell'}).text
print(price)
```

```python
price = shoe_info[3].find('span',{'class' : 'check_price_consumer'}).text
price
```

```python
name_list = []
price_list = []
url_list = []

for shoe in shoe_info :
    name = shoe.find('dt',{'class' : 'prd-name'}).text.strip()
    if shoe.find('span',{'class' : 'check_price_sell'}) is None :
        price = shoe.find('p',{'class' : 'sale'}).text
    else :
        price = shoe.find('span',{'class' : 'check_price_sell'}).text
    url = shoe.find('a').get('href')
    print("이름 : {}, 가격 : {}".format(name, price))
    name_list.append(name)
    price_list.append(price)
    url_list.append(url)
```

```python
df = pd.DataFrame({'이름' : name_list, '가격' : price_list, 'url' : url_list})
df.tail()
```
