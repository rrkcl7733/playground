```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
filename = 'humor.html'
html = ''
with open (filename, 'r', encoding ='euc-kr') as file:
    for line in file :
        html += line
html
```

```python
soup = BeautifulSoup(html, 'html')
soup
```

```python
a = soup.find('a',{'class' : 'li'}).text.strip()
a
```

```python
soup.find_all('a',{'class' : 'li'})
```

```python
title_list = []
```

```python
title = soup.find_all('a',{'class' : 'li'})
for post in title :
    a = post.text.strip()
    title_list.append(a)
title_list
```

```python
nick_list = soup.find_all('span',{'class' : 'hu_nick_txt'},limit=10)
```

```python
for nick in nick_list :
    print(nick.text.strip())
    
```

```python
df = pd.DataFrame({'title' : title_list, 'nick' : nick_list})
df
```
