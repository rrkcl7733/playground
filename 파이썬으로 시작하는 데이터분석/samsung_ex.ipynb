```python
df.reset_index(inplace = True)
# df['Date'] = df.index 가능
```

```python
df.rename(columns = {'Date' : '날짜', 'High' : '고가', 'Low' : '저가'})
```

```python
df.drop(columns = ['Date'])
```

```python
df['diff'] = df['Close'] - df['Open']
```

```python
df['Year'] = [d[0:4] for d in df['Date']]
df['Month'] = [d[5:7] for d in df['Date']]
df['Day'] = [d[8:10] for d in df['Date']]
df['Year'] = df['Year'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Day'] = df['Day'].astype(int)
df
```

```python
format = lambda x : '%.2f' % x
df.applymap(format)
```

```python
df['거래량'] = df['Volume'].apply(lambda x : "{:,}".format(x))
```

```python
volume_mean = df['Volume'].groupby(df['Month']).mean().round(2)
volume_mean
```
