```python
def getDate(data) :
    data = str(data)
    year = data[0:4]
    month = data[4:6]
    day = data[6:8]
    return year + '-' + month + '-' + day
```

```python
getDate(20210710)
```

```python
data['사용일자1'] = data['사용일자'].apply(getDate)  # apply 는 함수를 받아 모든 열이 적용시킴
data.tail()
```
